### Cleaning data

```python
df6=df5[df5["engine_capacity"]!="n/d"]
```

```python
df6.shape
```

```python
df7=df6.copy()
```

```python
#converting into float
df7["engine_capacity"]=df6["engine_capacity"].astype(int)
```

```python
df7["cylinder"].unique()
```

```python
df8=df7[df7["cylinder"]!="n"]
```

```python
df8.shape
```

```python
df9=df8.copy()
```

```python
df9["cylinder"]=df8["cylinder"].astype(int)
```

```python
df10=df9[df9["kw"]!="d"]
```

```python
df10.shape
```

```python
df11=df10.copy()
```

```python
df11["kw"]=df10["kw"].astype(int)
```

```python
df11["acceleration"].value_counts()
```

```python
df12=df11[df11["acceleration"]!="0"]
```

```python
df12.shape
```

```python
df13=df12.copy()
```

```python
df13["acceleration"]=df12["acceleration"].apply(lambda x: x.replace("(","").replace(")",""))
```

```python
df13.sample(10)
```

```python
def converting(x):
    tokens=x.split(" ") #splitting on " "
    if len (tokens)== 2:
        return (float(tokens[0])+float(tokens[1]))/2 #returning average values
    try:
        return float(x) #return float
    except:
        return None #ignore value that is not float
```

```python
df14=df13.copy()
```

```python
df14["acceleration"]=df13["acceleration"].apply(converting)
```

```python
df14.sample(10)
```

```python
df14["max_speed"]=df13["max_speed"].apply(lambda x:x.replace("(","").replace(" )","").replace("",""))
```

```python
df14.sample(10)
```

```python
df15=df14[df14["max_speed"]!="n/d"]
```

```python
def convert(x):
    tokens=x.split(" ") #splitting on " "
    if len (tokens)== 3:
        return (int(tokens[0])+int(tokens[2]))/2 #returning average values
    try:
        return float(x) #return float
    except:
        return None #ignore value that is not float
```

```python
df16=df15.copy()
```

```python
df16["max_speed"]=df15["max_speed"].apply(convert)
```

```python
df16.sample(10)
```

```python
df16.shape
```

```python
df17=df16[df16["lenght"]!="n/d"]
```

```python
df18=df17[df17["width"]!="n/d"]
```

```python
df19=df18[df18["height"]!="n/d"]
```

```python
df20=df19[df19["weight"]!="n/d"]
```

```python
df21=df20[df20["consump"]!="n/d"]
```

```python
df21.shape
```

```python
df22=df21.copy()
```

```python
df22["car_brand"]=df21["name"].apply(lambda x: x.split("/")[1].strip())
df22.sample(10)
```

```python
df23=df22.copy()
```

```python
df23["car_brand"]=df22["name"].apply(lambda x: x.split("/")[1].strip())
df23.sample(10)
```

```python
df24=df23.copy()
```

```python
df24["car_model"]=df23["name"].apply(lambda x: x.split("/")[2].strip())
```

```python
def conv(x):
    tokens=x.split("/") #splitting on "-"
    if len (tokens)==2:
        return (int(tokens[0])+int(tokens[1]))/2 #returning average values
    try:
        return float(x) #return float
    except:
        return None #ignore value that is not float
```

```python
df25=df24.copy()
```

```python
df25["weight"]=df24["weight"].apply(conv)
```

```python
df25.sample(10)
```

```python
df26=df25.copy()
```

```python
df26["fuel_consumption"]=df25["consump"].apply(lambda x: x.split(" ")[0].strip())
```

```python
df26.sample(10)
```

```python
data=df26.drop(["Unnamed: 0","name","consump"],axis=1)
```

```python
data_2 = data.reindex(columns=['car_brand','car_model', 'fuel', 'engine_capacity', 'cylinder', 'kw',
       'acceleration', 'max_speed', 'lenght', 'width', 'height', 'weight', 'fuel_consumption'])
```

```python
data_2.sample(10)
```

```python
#saving csv file
data_2.to_csv("clean_cars",index=False)
```
